```python
# =============================================================================
# YouTube Transcript Ingestion with Whisper
# =============================================================================
#
# This notebook implements an alternative transcript ingestion pipeline
# for the QA YouTube Chatbot project.
#
# The goal of this step is to convert YouTube videos into structured
# transcript data that can later be processed by the RAG pipeline.
#
# ---------------------------------------------------------------------------
# Pipeline Overview
# ---------------------------------------------------------------------------
#
# YouTube Video
#       ↓
# Download audio (yt-dlp)
#       ↓
# Transcribe speech using OpenAI Whisper
#       ↓
# Convert Whisper segments to normalized transcript schema
#       ↓
# Save transcript JSON files
#       ↓
# Build merged transcript dataset
#
# ---------------------------------------------------------------------------
# Output
# ---------------------------------------------------------------------------
#
# The notebook produces the following files:
#
# raw/t_01.json
# raw/t_02.json
# ...
# raw/all_transcripts.json
#
# Each transcript contains structured segments:
#
# {
#     "text": "...",
#     "start": seconds,
#     "duration": seconds,
#     "start_hhmmss": "HH:MM:SS",
#     "end_hhmmss": "HH:MM:SS"
# }
#
# ---------------------------------------------------------------------------
# Role in the Full RAG Pipeline
# ---------------------------------------------------------------------------
#
# This notebook corresponds to the **data ingestion stage** of the system.
#
# Full pipeline:
#
# 1. Transcript ingestion (this notebook or YouTube Transcript API)
# 2. Transcript chunking
# 3. Metadata enrichment
# 4. Vector database construction
# 5. Hybrid retrieval (FAISS + BM25)
# 6. Reranking
# 7. Context building
# 8. Answer generation with LLM
#
# ---------------------------------------------------------------------------
# Notes
# ---------------------------------------------------------------------------
#
# - Whisper transcription is slower but more robust when YouTube subtitles
#   are unavailable.
#
# - In the production pipeline of this project, transcripts are typically
#   obtained using the YouTube Transcript API for speed.
#
# - This notebook demonstrates a fallback approach based on speech-to-text.
#
# =============================================================================
```

```python
# STEP 0 — Install required dependencies
# This cell installs the external tools required for the transcription pipeline.
# - yt-dlp: downloads audio streams from YouTube videos
# - openai-whisper: speech-to-text model used to generate transcripts
#
# In a local environment these dependencies would normally be installed
# through requirements.txt, but in notebooks we install them directly.

!pip install -q yt-dlp openai-whisper
```

```python
# STEP 1 — Import required libraries
#
# json:
#     Used to serialize and save transcript outputs to disk.
#
# Path (from pathlib):
#     Provides an object-oriented way to manage filesystem paths.
#
# whisper:
#     OpenAI's speech-to-text model used to transcribe audio.
#
# yt_dlp:
#     Utility for downloading audio streams from YouTube videos.

import json
from pathlib import Path

import whisper
import yt_dlp
```

```python
# STEP 2 — Define the dataset
#
# VIDEO_IDS contains the YouTube videos that form the knowledge base
# of the RAG system.
#
# Each video will be:
# 1. Downloaded
# 2. Transcribed using Whisper
# 3. Converted into transcript segments
# 4. Saved as a JSON file (t_01.json, t_02.json, ...)
#
# These transcripts will later be chunked and indexed in the vector store.

VIDEO_IDS: list[str] = [
    # t_01
    "fIRM0bEZTpc",

    # t_02 ... t_08
    "MBtHiyY3FPo",
    "7F_HjxCLo6o",
    "JjSvBhc1gcg",
    "Qro3G8lKl0Y",
    "_nsLed3AHy8",
    "85UTQ6OWDFU",
    "KrHmB8nCU3U",
]
```

```python
# STEP 3 — Transcription configuration
#
# MODEL_NAME
#     Defines which Whisper model is used.
#     Larger models are more accurate but slower.
#
#     tiny  -> fastest, lowest accuracy
#     base
#     small
#     medium
#     large -> highest accuracy, slowest
#
# LANGUAGE
#     Language of the audio.
#     If set to None, Whisper performs automatic language detection.
#
# TASK
#     "transcribe" -> keep original language
#     "translate"  -> translate speech to English
#
# Directory structure
# -------------------
# AUDIO_DIR:
#     Stores downloaded audio files from YouTube.
#
# RAW_DIR:
#     Stores the generated transcript JSON files.

MODEL_NAME = "medium"   # tiny, base, small, medium, large
LANGUAGE = "es"         # usa None si quieres detección automática
TASK = "transcribe"     # o "translate"

BASE_DIR = Path("/content")
AUDIO_DIR = BASE_DIR / "audio"
RAW_DIR = BASE_DIR / "raw"

AUDIO_DIR.mkdir(exist_ok=True)
RAW_DIR.mkdir(exist_ok=True)
```

```python
# STEP 4 — Helper functions
#
# These utility functions support the transcription pipeline.
#
# hhmmss()
#     Converts timestamps (seconds) into human-readable HH:MM:SS format.
#
# find_downloaded_audio()
#     Checks whether the audio file for a video has already been downloaded.
#     This prevents unnecessary downloads when re-running the notebook.
#
# download_audio()
#     Uses yt-dlp to download the audio stream from a YouTube video.
#     Only the audio track is needed for Whisper transcription.

def hhmmss(seconds: float) -> str:
    total = int(round(seconds))
    h = total // 3600
    m = (total % 3600) // 60
    s = total % 60
    return f"{h:02d}:{m:02d}:{s:02d}"


def find_downloaded_audio(video_id: str) -> Path | None:
    matches = list(AUDIO_DIR.glob(f"{video_id}.*"))
    return matches[0] if matches else None


def download_audio(video_id: str) -> Path:
    url = f"https://www.youtube.com/watch?v={video_id}"
    outtmpl = str(AUDIO_DIR / f"{video_id}.%(ext)s")

    ydl_opts = {
        "format": "bestaudio/best",
        "outtmpl": outtmpl,
        "quiet": True,
        "noplaylist": True,
    }

    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([url])

    audio_path = find_downloaded_audio(video_id)
    if audio_path is None:
        raise FileNotFoundError(f"Audio file not found for {video_id}")

    return audio_path

# Convert Whisper transcription output into the project's normalized segment schema
def whisper_to_segments(result: dict) -> list[dict]:
    out = []

    for seg in result.get("segments", []):
        start = float(seg.get("start", 0.0))
        end = float(seg.get("end", start))
        text = str(seg.get("text", "")).replace("\n", " ").strip()

        if not text:
            continue

        out.append(
            {
                "text": text,
                "start": start,
                "duration": max(0.0, end - start),
                "start_hhmmss": hhmmss(start),
                "end_hhmmss": hhmmss(end),
            }
        )

    return out
```

```python
# STEP 5 — Load Whisper model
#
# The Whisper model is loaded once and reused for all videos.
#
# After loading, the model will be used to transcribe each downloaded audio file.

model = whisper.load_model(MODEL_NAME)
print(f"Whisper model loaded: {MODEL_NAME}")
```

```python
# STEP 6 — Transcribe audio using Whisper
#
# For each downloaded audio file:
# 1. The Whisper model processes the audio.
# 2. Speech is converted into text segments.
# 3. Each segment contains:
#       - text
#       - start timestamp
#       - end timestamp
#
# These segments are later converted into the project's normalized schema.

def process_video(video_id: str, index: int) -> dict:
    print(f"Processing {video_id}...")

    audio_path = download_audio(video_id)
    print(f"  Audio: {audio_path.name}")

    result = model.transcribe(
        str(audio_path),
        language=LANGUAGE,
        task=TASK,
        verbose=False,
    )

    segments = whisper_to_segments(result)

    payload = {
        "t_id": f"t_{index:02d}",
        "video_id": video_id,
        "language": result.get("language", LANGUAGE or "unknown"),
        "segments": segments,
    }

    out_file = RAW_DIR / f"t_{index:02d}.json"
    out_file.write_text(
        json.dumps(payload, ensure_ascii=False, indent=2),
        encoding="utf-8",
    )

    print(f"  Saved: {out_file.name} | segments: {len(segments)}")
    return payload
```

```python
# STEP 8 — Build transcript payload
#
# Each video transcript is stored as a structured JSON object:
#
# {
#     "t_id": "t_01",
#     "video_id": "...",
#     "language": "...",
#     "segments": [...]
# }
#
# This format is compatible with the rest of the ingestion pipeline.

all_payloads = []
errors = []

for i, video_id in enumerate(VIDEO_IDS, start=1):
    print(f"\n[{i}/{len(VIDEO_IDS)}] {video_id}")
    try:
        payload = process_video(video_id, i)
        all_payloads.append(payload)
    except Exception as e:
        print(f"  ERROR: {type(e).__name__}: {e}")
        errors.append({"video_id": video_id, "error": f"{type(e).__name__}: {e}"})

merged = {
    "total_videos": len(all_payloads),
    "videos": all_payloads,
    "errors": errors,
}

merged_path = RAW_DIR / "all_transcripts.json"
merged_path.write_text(
    json.dumps(merged, ensure_ascii=False, indent=2),
    encoding="utf-8",
)

print("\nDone.")
print(f"OK -> {merged_path}")
print(f"Success: {len(all_payloads)}/{len(VIDEO_IDS)}")
if errors:
    print("Errors:")
    for err in errors:
        print("-", err)
```
